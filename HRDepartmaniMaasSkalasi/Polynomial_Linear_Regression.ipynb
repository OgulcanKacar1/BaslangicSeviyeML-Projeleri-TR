```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("polynomial.csv",sep=";")

#* Bir adet polynomial regression nesenesi olusturması icin PolynomialFeatures fonksiyonunu cagiriyoruz.
#* Bu Fonksiyonu çağırırken polinomun derecesini (N) belirtiroyuz:
```

```python
 #* Veri Setimize Bir Bakalım
df
```

```python
plt.scatter(df["deneyim"],df["maas"],color="red")
plt.xlabel("Deneyim(Yil)")
plt.ylabel("Maas")
plt.savefig("1.png",dpi=300)
plt.legend()
plt.show()
```

```python
 #! Gördüğünüz gibi doğrusal bir yapıda dağılmıyor verilerimiz.
 #! Eğer biz bu veri setine linear regression uygularsak hiç ugun olmayan bir tahmin çizgisi görürüz.

reg = LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])


plt.scatter(df["deneyim"],df["maas"])
plt.xlabel("Deneyim (Yil)")
plt.ylabel("Maas")

xekseni = df["deneyim"]
yekseni = reg.predict(df[["deneyim"]])
plt.plot(xekseni,yekseni,color = "green",label = "linear regression")
plt.legend()
plt.show()
```

```python
polynomial_regression = PolynomialFeatures(degree =4)
x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
 #* regression model nesnemiz olan reg nesnemizi olusturup bunun fit metodunu cagırarak x_polynomial ve y
 #* yani regresyon modelimizi mevcut gerçek verilerle eğitiyoruz
reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head,color = "red",label="Polynomial Regression")
#* plt.plot(xekseni,yekseni,color="green",label="linear regression")
plt.legend()

#* Veri setimizi de noktalı olarak scatter edelim de görelim bakalım uymuş mu polynomial regression:
plt.scatter(df["deneyim"],df["maas"])
plt.show()
```
